**1. What is Simple Linear Regression?**
- A method used to model the relationship between a dependent variable (Y) and a single independent variable (X) by fitting a linear equation Y=mX+c.

**2. What are the key assumptions of Simple Linear Regression?**
- Linearity
- Independence of errors
- Homoscedasticity (constant variance)
- Normal distribution of residuals
- No significant outliers

**3. What does the coefficient m represent in the equation Y=mX+c?**
- The slope, showing how much Y changes for a unit change in X.

**4. What does the intercept c represent in the equation Y=mX+c?**
- The value of Y when X = 0; it’s the point where the line crosses the Y-axis.

**6. What is the purpose of the least squares method in Simple Linear Regression?**
- To minimize the sum of squared differences between observed and predicted values.

**7. How is the coefficient of determination (R²) interpreted in Simple Linear Regression?**
- Indicates the proportion of variance in Y explained by X (ranges from 0 to 1).

**8. What is Multiple Linear Regression?**
- A method to model the relationship between a dependent variable and two or more independent variables.

**9. What is the main difference between Simple and Multiple Linear Regression?**
- The main difference between Simple Linear Regression and Multiple Linear Regression lies in the number of independent (predictor) variables used to predict the dependent variable.

**10. What are the key assumptions of Multiple Linear Regression?**
- Linearity
- Independence of errors
- Homoscedasticity
- Normality of residuals
- No multicollinearity

**11. What is heteroscedasticity, and how does it affect the results of a Multiple Linear Regression model?**
- When the variance of residuals is not constant; it can lead to inefficient estimates and biased standard errors.

**12. How can you improve a Multiple Linear Regression model with high multicollinearity?**
- Remove or combine correlated predictors
- Use PCA or regularization techniques (Ridge, Lasso)

**13. What are some common techniques for transforming categorical variables for use in regression models?**
- One-hot encoding
- Label encoding
- Ordinal encoding (if order matters)

**14. What is the role of interaction terms in Multiple Linear Regression?**
- To capture the combined effect of two variables on the dependent variable (e.g., X1 , X2).

**15. How can the interpretation of intercept differ between Simple and Multiple Linear Regression?**
- Simple: value of Y when X = 0
- Multiple: value of Y when all Xs = 0 (may not be meaningful if Xs can't logically be 0)

**16. What is the significance of the slope in regression analysis, and how does it affect predictions?**
- In both Simple and Multiple Linear Regression, the slope (coefficient) represents the rate of change of the dependent variable Y with respect to an independent variable X.

**18. What are the limitations of using R² as a sole measure of model performance?**
- Key Limitations:
1. Does Not Indicate Causation
- A high R² does not mean that changes in the predictors cause changes in the response.
- It only shows correlation, not cause-and-effect.
2. Sensitive to Model Complexity
- Adding more predictors always increases R², even if they are not meaningful.
- This can lead to overfitting, where the model fits the training data too well but performs poorly on new data.
3. Does Not Reflect Model Bias
- A model with a high R² could still be biased if the assumptions (e.g., linearity, homoscedasticity) are violated.

**19. How would you interpret a large standard error for a regression coefficient?**
- In regression analysis, each coefficient (slope or intercept) comes with a standard error (SE) that measures the variability or uncertainty of that estimate.

**20. How can heteroscedasticity be identified in residual plots, and why is it important to address it?**
- Residual Plot: A scatter plot of residuals (Y-axis) vs. predicted values or an independent variable (X-axis)

**Signs of Heteroscedasticity:**

1. Funnel Shape / Cone Pattern:
- Residuals spread out or contract as fitted values increase.
- Looks like a triangle, fan, or hourglass.
2. Systematic Patterns:
- Curved or clustered shapes rather than a random cloud.

**21. What does it mean if a Multiple Linear Regression model has a high R² but low adjusted R²?**
- Suggests overfitting; too many variables that don't contribute meaningfully.

**22. Why is it important to scale variables in Multiple Linear Regression?**
- Especially for regularization and gradient descent; ensures variables contribute proportionally.

**23. What is polynomial regression?**
- Extension of linear regression where the relationship is modeled as an nth-degree polynomial.

**24. How does polynomial regression differ from linear regression?**
- Polynomial regression is an extension of linear regression that models nonlinear relationships by adding powers of the predictor variable(s). It provides more flexibility but must be used carefully to avoid overfitting.

**25. When is polynomial regression used?**
- When data shows a curvilinear trend that linear regression can’t capture.

**28. What are the limitations of polynomial regression?**
- While Polynomial Regression is powerful for modeling nonlinear relationships, it comes with several important drawbacks:

1. Overfitting Risk
- Higher-degree polynomials can fit the training data too closely, capturing noise instead of the true pattern.
- This leads to poor generalization on new/unseen data.
2. Extrapolation Issues
- Polynomial functions grow very fast at the extremes (especially with high degrees).
- Predictions outside the data range (extrapolation) can be wildly inaccurate or nonsensical.
3. Interpretability
- Coefficients in higher-degree models lack intuitive meaning.
- Harder to explain or justify to non-technical stakeholders.
4. Sensitivity to Outliers
- Like linear regression, polynomial regression is very sensitive to outliers, which can significantly distort the curve.

**29. What methods can be used to evaluate model fit when selecting the degree of a polynomial?**
- Choosing the right degree for polynomial regression is critical to avoid underfitting (too simple) or overfitting (too complex). Here are effective methods to guide model selection:
1. Train/Test Split Evaluation
- Split the data into training and testing sets.
- Train models of different degrees and compare their performance on the test set.

Metric to track:
- Mean Squared Error (MSE)
- Root Mean Squared Error (RMSE)
- R² Score (on test data)
2. Cross-Validation (e.g., K-Fold CV)
- Divide data into k folds, train on k-1, validate on the 1 remaining.
- Repeat for each fold and take the average error.
3. Adjusted R²
- Unlike regular R², adjusted R² penalizes unnecessary complexity.
- Helps detect if adding higher-degree terms truly improves the model.
4. AIC / BIC (Akaike/Bayesian Information Criterion)
- Measures model quality while penalizing complexity.
- Lower AIC/BIC = Better model (with balance between fit and simplicity)

**30. Why is visualization important in polynomial regression?**
- Visualization plays a crucial role in polynomial regression because it helps you:
1. Understand the Relationship Between Variables
- Polynomial regression models nonlinear relationships.
- A plot of the data with the regression curve allows you to see the shape of the fitted curve (e.g., U-shaped, S-shaped).
2. Detect Overfitting or Underfitting
- Underfitting: Curve is too flat or too simple
- Overfitting: Curve bends too much and follows noise in the data
- A visual plot makes these issues obvious by showing how well the curve aligns with actual data points.
3. Evaluate Model Fit Visually
- Even if metrics like R² or RMSE look good, the model might behave badly in certain regions.
- A visualization of the fitted curve shows whether it captures the true pattern or not.
4. Spot Anomalies and Outliers
- Outliers can distort the regression curve significantly, especially in polynomial regression.
- Plots help identify such points visually so you can decide whether to treat or remove them.
